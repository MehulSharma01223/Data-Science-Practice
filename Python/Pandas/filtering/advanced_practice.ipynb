```python
import pandas as pd
```

```python


data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Age": [20, 25, 22, 23, 21],
    "Marks": [50, 80, 75, 90, 60]
}

df = pd.DataFrame(data)

# TASK:
# 1. Marks > 70 wale students
# 2. Age < 23 AND Marks > 60
# 3. Sirf Name aur Marks print karo jaha Marks > 75
# 4. Marks 60–80 ke beech wale
```

```python
print(df[df["Marks"] > 70])
filtered = df[(df["Marks"] > 60 ) & (df["Age"] <23)]
print(filtered)
print(df[df["Marks"] > 75][["Name" ,"Marks"]])
print(df[df["Marks"].between(60,80)])
```

```python


data = {
    "Name": ["A", "B", "C", "D", "E", "F", "G"],
    "Age": [20, 25, 22, 23, 21, 24, 26],
    "Marks": [50, 80, 75, 90, 60, 55, 85],
    "City": ["Delhi", "Mumbai", "Delhi", "Pune", "Mumbai", "Delhi", "Pune"]
}

df = pd.DataFrame(data)
```

```python
#  Q 1 : Marks > 70
mask = df["Marks"] > 70
filtered = df[mask]
print(filtered)
print("Count : ", len(filtered))
```

```python
#  Q2. Age < 23 And Marks > 60
msk = (df["Marks"] > 60) & (df["Age"] < 23)
flt = df[msk]
print(flt)
print("Count : ", len(flt))
```

```python
#  Q3. Marks <60 OR Age > 24
msk = (df["Marks"] < 60) | (df["Age"] > 24)
flt = df[msk]
print(flt)
print("Count : ", len(flt))
```

```python
# Q4: Sirf Name aur Marks jaha Marks > 75
print(df[(df["Marks"] > 75)][["Name", "Marks"]])
```

```python
# Q5. Marks between 60 and 85
print(df[df["Marks"].between(60,85)])
```

```python
# Q6: City = "Delhi"
mask = df["City"] == "Delhi"
filterd = df[mask]
print(filterd)
print("Count : ",len(filterd))
```

```python
# Q7: City = Delhi AND Marks > 70
print(df[(df["City"] == "Delhi") & (df["Marks"] > 70)])
```

```python
# Q8: Names ["A", "D", "G"] select karo
filtered = df[df["Name"].isin(["A","D","G"])]
print(filtered)
```

```python
# Q9: Count students jinke Marks > 70
new = df["Marks"] > 70
count = df[new]
print("Count : ",len(count))
```

```python
# Q10: Marks < 60 walon ko 0 replace karo
df.loc[df["Marks"] < 60, "Marks"] = 0

print(df)
```

```python
# Q11: Age 21 se 25 ke beech + Marks > 60
print(df[(df["Age"].between(21,25)) & (df["Marks"] > 60)])
```

```python
# Q12: City Mumbai wale + sirf Name, Marks
print(df[df["City"] == "Mumbai"][["Name" , "Marks"]])
```

```python
# Q13: NOT condition (Marks <= 70)
mask = ~(df["Marks"] <= 70)
result = df[mask]

print(result)
```

```python
# Q14: Marks > 60 ke baad descending sort
filtered = df[df["Marks"] > 60]

result = filtered.sort_values(by="Marks", ascending=False)

print(result)
```

```python
# Q15: Top 3 highest marks wale students
top3 = df.sort_values(by="Marks", ascending=False).head(3)

print(top3)
```

```python
#  Sort values 
# 3 questions

data = {
    "Name": ["A", "B", "C", "D", "E", "F", "G"],
    "Age": [20, 25, 22, 23, 21, 24, 26],
    "Marks": [50, 80, 75, 90, 60, 55, 85],
    "City": ["Delhi", "Mumbai", "Delhi", "Pune", "Mumbai", "Delhi", "Pune"]
}

df = pd.DataFrame(data)
```

```python
# Q4: Marks ascending sort
sot = df.sort_values(by= "Marks",ascending= True)
print(sot)
```

```python
# Q5: Marks descending sort
sot = df.sort_values(by= "Marks",ascending= False)
print(sot)
```

```python
# Q6: City + Marks dono se sort
sot = df.sort_values(by= ["Marks","City"] ,ascending= True)
print(sot)
```
