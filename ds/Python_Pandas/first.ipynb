## Make a series using list

```python
country = ['India','Nepal','USA','Shrilanka','Bhutan']
pd.Series(country)
```

```python
# Intergers 
runs = [ 34,54,66,77,33,13]
pd.Series(runs)
```

```python
# Custom index
subjects= [ 'marks ', 'english ','science','hindi']
marks = [ 34,54,66,77]
pd.Series(marks, index = subjects)
```

```python
# setting a name
marks = pd.Series(marks,index= subjects,name = 'Mehul Sharma')
marks
```

```python
# Series from dict
marks = {
    'marks': 65,
    'english':74,
    'science': 94,
    'hindi': 78,
}
marks = pd.Series(marks,name = 'mehul ke Marks')
marks
```

## Series Attribues

```python
# (6) values
type(marks.values)
```
