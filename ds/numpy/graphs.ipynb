```python
import numpy as np
```

## Ploting Graphs

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,100)
y = x
plt.plot(x,y)
```

```python
# y = x^2
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y)
```

```python
# sinx
x = np.linspace(-10,10,100)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = np.linspace(-10,10,100)
y = x*(np.log(x))
plt.plot(x,y)
```

```python
x = np.linspace(-10,10,100)
y = np.log10(x)
plt.plot(x,y)
```

```python
x = np.linspace(-10,10,100)
y = 1/np.sin(x)
plt.plot(x,y)
```

```python
x = np.linspace(-10,10,100)
y = 1/ (1 + np.exp(-x))
plt.plot(x,y)
```

```python
x = np.linspace(-10,10000,1000000)
y = 1/np.cos(x)
plt.plot(x,y)
```

```python
x = np.linspace(-10,10,100)
y = x*x*x
plt.plot(x,y)
```
