```python
import numpy as ln
```

# Create a 2D array (5×5) #

```python
a = [3,5,4,22,55,6,7,8,5,34,65,67,55,66,345,34,0,4,45,45,56,67,56,32,56]
aar = ln.array(a)
s = aar.reshape(5,5)
print(s)
```

## Extract second row and third column ##

```python
print(s[1])
print(s[2])
```

##  Find row-wise mean (axis=1)

```python
d = s.mean(axis = 1)
print(d)
```

# Find column-wise sum (axis=0)

```python
e = s.sum(axis = 0)
print(e)
```

 ## Filter elements greater than 50 #

```python
g = s [s > 50]
print(g)
```

## Replace values less than 20 with 0 #

```python
s[ s < 20] = 0
print(s)
```

```python
l = s.ravel()
print(l)
```

```python
l = ln.sort(s,axis = None)
print(l[::-1])
```

#  Find index of maximum value

```python
print(l.max())
```

## Flatten a 2D array

```python
flt = s.ravel()
print(flt)
```
