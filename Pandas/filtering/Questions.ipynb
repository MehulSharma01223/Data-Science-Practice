```python
import pandas as pd

data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Age": [20, 25, 22, 23, 21],
    "Marks": [50, 80, 75, 90, 60]
}

df = pd.DataFrame(data)

# TASK:
# 1. Marks > 70 wale students
# 2. Age < 23 AND Marks > 60
# 3. Sirf Name aur Marks print karo jaha Marks > 75
# 4. Marks 60–80 ke beech wale
```

```python
print(df[df["Marks"] > 70])
filtered = df[(df["Marks"] > 60 ) & (df["Age"] <23)]
print(filtered)
print(df[df["Marks"] > 75][["Name" ,"Marks"]])
print(df[df["Marks"].between(60,80)])
```
